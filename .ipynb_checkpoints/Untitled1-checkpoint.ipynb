```python
import numpy as np
import pandas as pd
from pandas import read_excel

import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = read_excel('Student Survey - July.xlsx', sheet_name= 'net_0_Friends')
df = df.rename(columns={'Source': 'Participant-ID', 'Target': 'Target'})
df2 = read_excel('Student Survey - July.xlsx', sheet_name='participantsNov')
df3 = read_excel('Student Survey - July.xlsx', sheet_name='responses')
df3 = df3[df3['Status'].isin(['completed'])]

df4 = pd.merge(df, df2, on=['Participant-ID'])
df5 = pd.merge(df4, df3, on=['Participant-ID'])

df5 = df5[['Participant-ID', 'Target', 'House', 'CompleteYears']]
```

```python
house = ['Vanguard', 'Griffin', 'Phoenix', 'Falcon', 'Redwood', 'Astral']

no_of_student = []

for index in house:
    no_of_student.append(df5.loc[df5['House'] == index, 'House'].count())
```
