# Principal dashboard

```python
import numpy as np
import pandas as pd
from pandas import read_excel

import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data source

# survey result  
df = read_excel('Student Survey - July.xlsx', sheet_name='responses')

# students attributes
df2 = read_excel('Student Survey - July.xlsx', sheet_name='participantsNov')

# inner merge on participant ID
df3 = pd.merge(df, df2, on=['Participant-ID'])

# data cleaning
# select only 'completed'
df3 = df3[df3['Status'].isin(['completed'])]
```

## Landing page

```python
print('Total student participants :', len(df3))
```

```python
year = []
for i in range (11):
    year.append(i)

student_by_year = []
for index in year:
    student_by_year.append(df3.loc[df3['CompleteYears'] == index, 'CompleteYears'].count())
```

```python
plt.bar(year, student_by_year)
plt.xlabel('Year')
plt.ylabel('No. of student')
plt.title('No. of student participated in the survey')
plt.show()
```

```python
house = ['Vanguard', 'Griffin', 'Phoenix', 'Falcon', 'Redwood', 'Astral']

student_by_house = []
for index in house:
    student_by_house.append(df3.loc[df3['House'] == index, 'House'].count())
```

```python
plt.bar(house, student_by_house)
plt.xlabel('Year')
plt.ylabel('No. of student')
plt.title('No. of student participated in the survey')
plt.show()
```

## View by year

### 1. What is their academic result and attandance?

```python
year = []
for i in range (11):
    year.append(i)

attandance_by_year = []
academic_by_year = []
for index in year:
    attandance_by_year.append(df3.loc[df3['CompleteYears'] == index, 'Attendance'].mean()) 
    academic_by_year.append(df3.loc[df3['CompleteYears'] == index, 'Perc_Academic'].mean()) 
```

```python
plot_data = np.transpose([academic_by_year, attandance_by_year])
working_frame = pd.DataFrame(plot_data, columns=['Academic', 'Attandance'])
```

```python
plot = working_frame.plot(kind='bar', stacked=False)
plot.set_xticklabels(year, rotation=0)
plot.set_xlabel('Year')
plot.set_ylabel('Academic result / attandance')
plot.set_title('Average academic result and attandance of students of each year')
plt.show()
```

### 2 Which language does they speak?

```python
year = []
for i in range (11):
    year.append(i)

english_percent_by_year = []
nonenglish_percent_by_year = []
for index in year:
    nonenglish_percent_by_year.append(len(df3.loc[(df3['language'] == 1) & (df3['CompleteYears'] == index)]) / len(df3[df3['CompleteYears'] == index]))
    english_percent_by_year.append(len(df3.loc[(df3['language'] == 0) & (df3['CompleteYears'] == index)]) / len(df3[df3['CompleteYears'] == index]))    
```

```python
data = np.transpose([english_percent_by_year, nonenglish_percent_by_year])
working_frame = pd.DataFrame(data, columns=['English speaking', 'Non-english speaking'])
plot = working_frame.plot(kind='bar', stacked=True)
plot.set_xticklabels(year, rotation=0)
plot.set_xlabel('Year')
plot.set_ylabel('Percentage %')
plot.set_title('Composition of English and Non-english speaking student')
plt.show()
```

#### 2.1 Is language affecting academic result?

```python
year = []
for i in range (11):
    year.append(i)
    
english_academic_year = []
nonenglish_academic_year = []
for index in year:
    english_academic_year.append(df3.loc[(df3['language'] == 0) & (df3['CompleteYears'] == index), 'Perc_Academic'].mean())
    nonenglish_academic_year.append(df3.loc[(df3['language'] == 1) & (df3['CompleteYears'] == index), 'Perc_Academic'].mean())
```

```python
data = np.transpose([english_academic_year, nonenglish_academic_year])
working_frame = pd.DataFrame(data, columns=['English speaking', 'Non-english speaking'])
plot = working_frame.plot(kind='bar', stacked=False)
plot.set_xticklabels(year, rotation=0)
plot.set_xlabel('Year')
plot.set_ylabel('Academic result')
plot.set_title('Average academic result of English and Non-English speaking student')
plt.show()
```

```python
house = ['Vanguard', 'Griffin', 'Phoenix', 'Falcon', 'Redwood', 'Astral']
sna = ['net_0_Friends', 'net_1_Influential', 'net_2_Feedback', 'net_3_MoreTime', 'net_4_Advice', 'net_5_Disrespect']
    
house_html = house[0] # user select a house
sna_html = sna[2] # user select a score to view
```

```python
df = read_excel('Student Survey - July.xlsx', sheet_name=sna_html)
df = df.rename(columns={'Source': 'Participant-ID', 'Target': 'Target'})
df2 = read_excel('Student Survey - July.xlsx', sheet_name='participantsNov')
df3 = pd.merge(df, df2, on=['Participant-ID'])
df3 = df3[['Participant-ID', 'Target', 'House']][df3['House'] == house_html]
```

```python
# Create a graph from the data
G = nx.from_pandas_edgelist(df3, source='Participant-ID', target='Target')

# Visualize the graph
pos = nx.spring_layout(G)
labels = {node: node for node in G.nodes()}

# Create a figure and axis objects
fig, ax = plt.subplots()

# Draw the graph on the axis
nx.draw(G, pos, with_labels=True, labels=labels, node_size=500, node_color='skyblue', font_size=8, ax=ax)

# Show the plot
plt.show()
```
